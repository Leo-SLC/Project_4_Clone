```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

import pandas as pd
loans_df = pd.read_csv("https://raw.githubusercontent.com/Jleavitt0/Final_Project/main/Loan_default.csv")
loans_df.head()
```

```python
loans_df.insert(0,'#',range(len(loans_df)))
loans_df.head()
```

```python
loans_df = loans_df.drop(columns=['LoanID','Education','EmploymentType','MaritalStatus', 'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner'])
loans_df
```

```python
loans_df.nunique()
```

```python
loan_counts = loans_df.Default.value_counts()
loan_counts
```

```python
y = loans_df['Default'].values

X = loans_df.drop('Default',1).values

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#X_train_scaled

number_of_input_features = len(X_train[0])

hidden_layer_1 = 10

hidden_layer_2 = 10

nn = tf.keras.models.Sequential()

# First hidden layer

nn.add(tf.keras.layers.Dense(units=hidden_layer_1, input_dim=number_of_input_features, activation="relu"))

# Second hidden layer

nn.add(tf.keras.layers.Dense(units=hidden_layer_2, input_dim=number_of_input_features, activation="relu"))

# Output layer

nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
nn.fit(X_train_scaled, y_train, epochs=20)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file

nn.save("loan_model.h5")
```
